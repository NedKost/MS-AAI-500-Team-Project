```python
# Read in full Motor Vehicle Collision Dataset
mvc_df = pd.read_csv('Motor_Vehicle_Collisions_-_Crashes.csv')
```

```python
# Set key Filter words limit the dataset
filter_words = 'Alcohol|Speed|Drugs'

# Filter out rows based on Contributing Factor for Vehicle 1 (first get rid of nulls, then filter on keywords)
cf1 = mvc_df[~mvc_df['CONTRIBUTING FACTOR VEHICLE 1'].isnull()]
cf1 = cf1[cf1['CONTRIBUTING FACTOR VEHICLE 1'].str.contains(filter_words)]

# Filter out rows based on Contributing Factor for Vehicle 2 (first get rid of nulls, then filter on keywords)
cf2 = mvc_df[~mvc_df['CONTRIBUTING FACTOR VEHICLE 2'].isnull()]
cf2 = cf2[cf2['CONTRIBUTING FACTOR VEHICLE 2'].str.contains(filter_words)]

# Filter out rows based on Contributing Factor for Vehicle 3 (first get rid of nulls, then filter on keywords)
cf3 = mvc_df[~mvc_df['CONTRIBUTING FACTOR VEHICLE 3'].isnull()]
cf3 = cf3[cf3['CONTRIBUTING FACTOR VEHICLE 3'].str.contains(filter_words)]

# Filter out rows based on Contributing Factor for Vehicle 4 (first get rid of nulls, then filter on keywords)
cf4 = mvc_df[~mvc_df['CONTRIBUTING FACTOR VEHICLE 4'].isnull()]
cf4 = cf4[cf4['CONTRIBUTING FACTOR VEHICLE 4'].str.contains(filter_words)]

# Filter out rows based on Contributing Factor for Vehicle 5 (first get rid of nulls, then filter on keywords)
cf5 = mvc_df[~mvc_df['CONTRIBUTING FACTOR VEHICLE 5'].isnull()]
cf5 = cf5[cf5['CONTRIBUTING FACTOR VEHICLE 5'].str.contains(filter_words)]
```
